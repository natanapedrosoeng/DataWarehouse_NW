## Analisando tipo de dados e dados nulos

## Ordens de pedidos

```python
order = pd.read_csv('orders.csv', sep = ';')
order_df = pd.DataFrame(order)
display(order_df)
order_df.info()
```

### Identificando dados nulos

```python
nulos = order_df[order_df['ship_region'].isnull()]
display(nulos)
nulos.info()
print(nulos.isnull().sum())
```
